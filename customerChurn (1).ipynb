```python

from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
#make a copy of dataset
train_df = df.copy()
#seperate target variables from independent variable
y = df['churn']
x = df.drop(columns = ['id' , 'churn'])
print(x.shape)
print(y.shape)
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.25 , random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = RandomForestClassifier(n_estimators=1000) #add parameter to the model
model.fit(x_train ,y_train)
```

```python
predictions = model.predict(x_test)
tn , fn , fp , tp = metrics.confusion_matrix(y_test , predictions).ravel()
```

```python
y_test.value_counts()
```

```python
print(f"true positives : {tp}")
print(f"false positives : {fp}")
print(f"true negatives : {tn}")
print(f"true negatives : {fn}\n")
print(f"accuracy : {metrics.accuracy_score(y_test , predictions)}")
print(f"precision : {metrics.precision_score(y_test , predictions)}")
print(f"recall : {metrics.recall_score(y_test , predictions)}")
```
